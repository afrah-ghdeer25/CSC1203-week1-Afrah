```python
import os


base_dir = "lab_data"
if not os.path.exists(base_dir):
    os.makedirs(base_dir)
    print(f"Folder '{base_dir}' created.")
else:
    print(f"Folder '{base_dir}' already exists.")


subfolders = ['input', 'output', 'logs']
for sub in subfolders:
    path = os.path.join(base_dir, sub)
    if not os.path.exists(path):
        os.makedirs(path)


print("\nContents of 'lab_data':")
for item in os.listdir(base_dir):
    full_path = os.path.join(base_dir, item)
    if os.path.isdir(full_path):
        print(f"{item} is a directory")
    else:
        print(f"{item} is a file")


old_path = os.path.join(base_dir, "logs")
new_path = os.path.join(base_dir, "history")
if os.path.exists(old_path):
    os.rename(old_path, new_path)
    print("\nRenamed 'logs' to 'history'.")
```

```python

notes_path = os.path.join("lab_data", "input", "notes.txt")
lines = [
    "This is the first line.\n",
    "Learning Python OS module.\n",
    "Working with directories is easy.\n",
    "Jupyter Notebook is great for labs.\n",
    "This is the fifth line.\n"
]

with open(notes_path, "w") as f:
    f.writelines(lines)


with open(notes_path, "r") as f:
    content = f.readlines()
    print(f"\nTotal number of lines in notes.txt: {len(content)}")


summary_path = os.path.join("lab_data", "output", "summary.txt")
with open(notes_path, "r") as f_in:
    data = f_in.read()
    with open(summary_path, "w") as f_out:
        f_out.write(data)
print("Contents copied to summary.txt.")
```
